## Задание 6 (1 балл)

Модифицируйте функцию, осуществляющую метод градиентного спуска с семинара: добавьте опцию останавливать метод, если абсолютное значение производной функции в точке < некоторого $\delta$ (delta добавьте в аргументы функции).

```python
def gradient_descent(func, x_symbol, x_start, learning_rate, epsilon, num_iterations, delta):
    x_curr = x_start
    df_x = sp.diff(func, x_symbol)

    trace = []
    trace.append(x_curr)

    for i in range(num_iterations):
        grad_value = df_x.subs(x_symbol, x_curr)
        if abs(grad_value) < delta:
            return x_curr, trace

        x_new = x_curr - learning_rate * grad_value
        trace.append(x_new)

        if abs(x_new - x_curr) < epsilon:
            return x_curr, trace

        x_curr = x_new

    return x_curr, trace
```

## Задание 8 (1.5 балла)

Реализуйте [метод Ньютона](https://en.wikipedia.org/wiki/Newton%27s_method_in_optimization) (метод второго порядка) для поиска минимума функции.

Про метод ещё можно почитать [здесь](https://math.semestr.ru/optim/minimum.php).

```python
def newton_method(func, x_symbol, x_start, epsilon, num_iterations, delta):
    x_curr = x_start
    df_x = sp.diff(func, x_symbol)
    d2f_x = sp.diff(df_x, x_symbol)
    trace = []
    trace.append(x_curr)

    for i in range(num_iterations):
        grad_value = df_x.subs(x_symbol, x_curr)
        hessian_value = d2f_x.subs(x_symbol, x_curr)

        if abs(hessian_value) < 1e-10:
            print("Вторая производная близка к нулю, метод Ньютона не применим.")
            return x_curr, trace

        if abs(grad_value) < delta:
            return x_curr, trace

        x_new = x_curr - grad_value / hessian_value
        trace.append(x_new)

        if abs(x_new - x_curr) < epsilon:
            return x_new, trace

        x_curr = x_new

    return x_curr, trace
```

## Задание 11 (1 балл)

Реализуйте оптимизацию метода градиентного спуска - Momentum. Необходимые выкладки вы можете подсмотреть [здесь](https://education.yandex.ru/handbook/ml/article/optimizaciya-v-ml).

```python
def gradient_descent_momentum(func, x_symbol, x_start, learning_rate, beta, epsilon, num_iterations, delta):
    x_curr = x_start
    df_x = sp.diff(func, x_symbol)
    velocity = 0
    trace = [x_curr]

    for i in range(num_iterations):
        grad_value = df_x.subs(x_symbol, x_curr)

        if abs(grad_value) < delta:
            return x_curr, trace

        velocity = beta * velocity - learning_rate * grad_value

        x_new = x_curr + velocity
        trace.append(x_new)

        if abs(x_new - x_curr) < epsilon:
            return x_new, trace

        x_curr = x_new

    return x_curr, trace
```

## Задание 12 (0.5 балла)

Проверьте, решает ли Momentum GD проблему поиска начального приближения для функций из задания 1?

```python
functions = {
    "f(x)": (x**2 - 30*x + 1, 15, (0, 30)),
    "g(x)": (-x**3 - 7*x**2 + 5*x + 17, -5, (-10, 10)),
    "h(x)": (x**2 - 1/x, 1, (0.1, 10)),
    "j(x)": ((1/x)**5 + x**17 + 8, 0.5, (0.1, 2))
}

learning_rate = 0.1
momentum = 0.9
epsilon = 1e-6
delta = 1e-5
num_iterations = 1000

results_momentum = {}

for name, (func, x_true, x_range) in functions.items():
    x_start = np.random.uniform(*x_range)
    minimum_x, trace = gradient_descent_momentum(func, x, x_start, learning_rate, momentum, epsilon, num_iterations, delta)
    error = abs(minimum_x - x_true)
    results_momentum[name] = {
        "x_start": x_start,
        "minimum_x": minimum_x,
        "error": error,
        "iterations": len(trace)
    }
```

```python
for name, result in results_momentum.items():
    print(f"Results for {name}:")
    print(f"  x_start = {float(result['x_start']):.2f}, minimum_x = {float(result['minimum_x']):.2f}, error = {float(result['error']):.2e}")
    print(f"  Iterations = {result['iterations']}")
    print()
```
